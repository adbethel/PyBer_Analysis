```python
%matplotlib inline
```

```python
#Import dependencies
import matplotlib.pyplot as plt
```

```python
#Set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]

#Set the y_axis to a list of floats as the total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
#Create the plot
plt.plot(x_axis, y_axis)
```

```python
#Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
#create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
#Create the plot
plt.plot(x_axis, y_axis)
plt.show()
```

```python
#Create the plot and add a label for the legend.
plt.plot(x_axis, y_axis,marker = "*", color = "green", linewidth= 2, label= "Boston")
#Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
#Set the limit between 0 and 45
plt.ylim(0, 45)
#Create a title
plt.title("PyBer Fare by Month")
#Add a grid
plt.grid()
#Add a legend
plt.legend()
```

```python
#Create the plot
plt.bar(x_axis, y_axis, color = "green", label="Boston")
#Create labels for x and y axes
plt.xlabel("Date")
plt.ylabel("Fare($)")
#Create a title
plt.title("PyBer Fare by Month")
#Add legend
plt.legend()
```

```python
#Create the plot
plt.barh(x_axis, y_axis)
```

```python
plt.barh(y_axis, x_axis)
```

```python
plt.barh(x_axis, y_axis, color= "magenta")
plt.gca().invert_yaxis()
plt.legend("Boston")
plt.title("High School Music")
plt.xlabel("Number")
plt.ylabel("Month")
```

```python
#Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis)
```

```python
#Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.barh(x_axis, y_axis, color="cyan")
ax.legend("Chicago")
ax.set_title("City Bar chart")
ax.set_xlabel("Fare($)")
ax.set_ylabel("Date")
```

```python
plt.plot(y_axis, x_axis, "o", color = "Red")
plt.legend("Chicago")
plt.title("PyBer Fare by Month")
plt.xlabel("Fares($)")
plt.ylabel("Date")
plt.xlim(0, 50)
```

```python
plt.scatter(x_axis, y_axis)
```

```python
plt.scatter(x_axis, y_axis, s=y_axis)
```

```python
y_axis_larger = []
for data in y_axis:
    y_axis_larger.append(data*3)
plt.scatter(x_axis, y_axis, s=y_axis_larger)
```
