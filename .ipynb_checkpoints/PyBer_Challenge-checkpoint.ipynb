```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
```

```python
city_data_df = pd.read_csv("PyBer_Resources/city_data.csv")
city_data_df.head(10)
```

```python
ride_data_df = pd.read_csv("Pyber_Resources/ride_data.csv")
ride_data_df.head(10)
```

```python
pyber_data_df = pd.merge(ride_data_df, city_data_df, how="left", on=["city", "city"])
pyber_data_df.head()
```

```python
#Get the total rides for each city type
city_rides = pyber_data_df.groupby("type").count().ride_id
city_rides
```

```python
#Get total drivers by city type
city_drivers = city_data_df.groupby("type").sum().driver_count
city_drivers
```

```python
#Get the total fares by city
city_fares = pyber_data_df.groupby("type").sum().fare
city_fares
```

```python
#average fare per ride by city type
avg_fare_by_city = pyber_data_df.groupby("type").mean().fare
avg_fare_by_city
```

```python
avg_fare_driver_by_city = pyber_data_df.groupby("type").sum().fare/city_drivers
avg_fare_driver_by_city
```

```python
pyber_summary_df = pd.DataFrame(
                {"Total Rides": city_rides,
                "Total Drivers": city_drivers,
                "Total Fares": city_fares,
                "Average Fare per Ride": avg_fare_by_city,
                "Average Fare per Driver": avg_fare_driver_by_city}
                )
pyber_summary_df
```

```python
pyber_summary_df.index.name = None
pyber_summary_df
```

```python
pyber_summary_df['Total Rides'] = pyber_summary_df['Total Rides'].map('{:,}'.format)
pyber_summary_df['Total Drivers'] = pyber_summary_df['Total Drivers'].map('{:,}'.format)
pyber_summary_df['Total Fares'] = pyber_summary_df['Total Fares'].map('${:,.2f}'.format)
pyber_summary_df['Average Fare per Ride'] = pyber_summary_df['Average Fare per Ride'].map('${:,.2f}'.format)
pyber_summary_df['Average Fare per Driver'] = pyber_summary_df['Average Fare per Driver'].map('${:,.2f}'.format)

pyber_summary_df
```

```python
pyber_data_df.head(4)
```

```python
fare_by_city_df = pyber_data_df.groupby(["type", "date"]).sum().fare.to_frame()
fare_by_city_df
```

```python
fare_by_city_df = fare_by_city_df.reset_index()
fare_by_city_df.head()
```

```python
fare_by_city_date_df = fare_by_city_df.pivot(index="date", columns="type", values="fare")
fare_by_city_date_df.head()
```

```python
fare_dates_df = fare_by_city_date_df.loc["2019-01-1":"2019-04-29"]
fare_dates_df
```

```python
type(fare_dates_df.index)

fare_dates_df.index = pd.to_datetime(fare_dates_df.index)
fare_dates_df.head()
```

```python
fare_dates_df.info()
```

```python
fare_dates_df = fare_dates_df.resample("W").sum()
fare_dates_df
```

```python
# Import the style from Matplotlib.
from matplotlib import style
# Use the graph style fivethirtyeight.
style.use('fivethirtyeight')

fare_dates_df.plot(figsize=(18, 5),)
plt.xlabel("")
plt.ylabel("Fare ($USD)")
plt.title("Total Fare by City Type")
plt.savefig("Analysis/PyBer_summary_chart.png")
```
